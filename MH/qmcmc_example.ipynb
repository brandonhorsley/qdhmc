```python
from qmcmc import IsingMH
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = 5
r = 5
t = 0.5
reps = 5
burn_in = 100

cs_5 = []
qs_5 = []

for rep in range(reps):
    if rep % 10 == 0:
        print(rep)
    js_1d = np.random.normal(0, 1, s)
    hs = np.random.normal(0, 1, s)


    test_c = IsingMH(s, js_1d, hs, r, t, "classical")
    test_q = IsingMH(s, js_1d, hs, r, t, "quantum")

    _, _, _, a_c, rs_c = test_c.run_mcmc(100, burn_in)
    _, _, _, a_q, rs_q = test_q.run_mcmc(100, burn_in)
    cs_5.append(-rs_c[0])
    qs_5.append(-rs_q[0])

s = 10

cs_10 = []
qs_10 = []

for rep in range(reps):
    if rep % 10 == 0:
        print(rep)
    js_1d = np.random.normal(0, 1, s)
    hs = np.random.normal(0, 1, s)


    test_c = IsingMH(s, js_1d, hs, r, t, "classical")
    test_q = IsingMH(s, js_1d, hs, r, t, "quantum")

    _, _, _, a_c, rs_c = test_c.run_mcmc(100, burn_in)
    _, _, _, a_q, rs_q = test_q.run_mcmc(100, burn_in)
    cs_10.append(-rs_c[0])
    qs_10.append(-rs_q[0])
    
s = 15

cs_15 = []
qs_15 = []

for rep in range(reps):
    if rep % 10 == 0:
        print(rep)
    js_1d = np.random.normal(0, 1, s)
    hs = np.random.normal(0, 1, s)


    test_c = IsingMH(s, js_1d, hs, r, t, "classical")
    test_q = IsingMH(s, js_1d, hs, r, t, "quantum")

    _, _, _, a_c, rs_c = test_c.run_mcmc(100, burn_in)
    _, _, _, a_q, rs_q = test_q.run_mcmc(100, burn_in)
    cs_15.append(-rs_c[0])
    qs_15.append(-rs_q[0])
```

```python
fig, ax = plt.subplots(figsize=(14, 6))
x = np.arange(100)
cs = np.array(cs_5)
qs = np.array(qs_5)

cs1 = np.squeeze(cs)
qs1 = np.squeeze(qs)

cs_mean = np.mean(cs1, axis=0)
qs_mean = np.mean(qs1, axis=0)

cs_ci = 1.96 * np.std(cs1, axis=0)/np.sqrt(100)
qs_ci = 1.96 * np.std(qs1, axis=0)/np.sqrt(100)

ax.plot(cs_mean, label='Classical')
ax.plot(qs_mean, label='Quantum')

ax.fill_between(x, (cs_mean - cs_ci), (cs_mean + cs_ci), color='b', alpha=.1)
ax.fill_between(x, (qs_mean - qs_ci), (qs_mean + qs_ci), color='orange', alpha=.1)

plt.title("Size 5 System")
plt.xlabel("Iteration")
plt.ylabel("Energy")
plt.legend()
plt.show()
```
